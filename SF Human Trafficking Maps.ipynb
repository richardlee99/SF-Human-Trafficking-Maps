```python
import pandas as pd
import geopandas as gpd
import math
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
crimes = pd.read_csv('Police_Department_Incident_Reports__2018_to_Present.csv', encoding='latin-1')
```

```python
crimes.dropna(subset=['Latitude', 'Longitude', 'Police District'], inplace=True)
```

```python
crimes = crimes[crimes['Incident Category'].isin(['Missing Person', 'Human Trafficking (A)', 'Human Trafficking (B)'])]
```

```python
crimes['Police District'] = crimes['Police District'].str.upper()
```

```python
cluster_map = folium.Map(location=[37.7879,-122.4480],tiles='cartodbpositron', zoom_start=10)
```

```python
mc = MarkerCluster()
for idx, row in crimes.iterrows():
    if not math.isnan(row['Longitude']) and not math.isnan(row['Latitude']):
        mc.add_child(Marker([row['Latitude'], row['Longitude']]))                    
```

```python
cluster_map.add_child(mc)
```

```python
heat_map = folium.Map(location=[37.7879,-122.4480],tiles='cartodbpositron', zoom_start=10)
```

```python
HeatMap(data=crimes[['Latitude', 'Longitude']], radius=10).add_to(heat_map)
```

```python
sf_districts = gpd.read_file('geo_export_777eea23-fff4-459a-b103-49cb7069ec64.shp')
```

```python
districts = sf_districts[['district', 'geometry']].set_index('district')
```

```python
district_sum = crimes['Police District'].value_counts()
```

```python
folium_map = folium.Map(location=[37.7879,-122.4480],tiles='cartodbpositron', zoom_start=10)
```

```python
Choropleth(geo_data=districts.__geo_interface__, data=district_sum, key_on="feature.id", fill_color='YlGnBu',legend_name='Human Trafficking in SF since 2018').add_to(folium_map)
```
